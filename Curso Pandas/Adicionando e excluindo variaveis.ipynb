## CRIANDO NOVAS VARIÁVEIS

```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/aluguel_residencial.csv', sep=';')
dados
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados
```

```python
dados['Valor M2'] = (dados['Valor'] / dados['Area']).round(2)
dados
```

```python
dados['Valor Bruto M2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.info
```

## Separando imóveis residenciais

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo Variáveis

```python
# Criando um DataFrame para testes
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor M2', 'Valor Bruto', 'Valor Bruto M2']])
```

```python
# Métodos de exclusão de variáveis
```

```python
# del DataFrame['Variável']
del dados_aux['Valor Bruto']
```

```python
#DataFrame.pop('Variável')
dados_aux.pop('Valor Bruto M2')
```

```python
#DataFrame.drop(['Variável 1', 'Variável 2'], axis = 1)  pode usar nas linhas ou colunas, por isso o parametro "axis"
dados.drop(['Valor Bruto', 'Valor Bruto', 'Tipo Agregado', 'Valor Bruto M2'], axis= 1, inplace= True)
```
