## Tratamento de dados faltantes, parte 2

```python
selecao_apto = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
dados_apto = dados[selecao_apto]
dados_apto
```

```python
#retirando o "pedaço de DF acima do DF original"
a = dados.shape[0]
#O '~' inverte a seleção feita, ou seja, cria um DF sem a seleção escolhida
dados = dados[~selecao_apto]
b = dados.shape[0]


```

```python
dados = dados.fillna({'Condominio': 0, 'IPTU' : 0} )
```
